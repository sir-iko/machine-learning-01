```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('garments_worker_productivity.csv')
dataset = dataset.drop(['date', 'wip', 'quarter'], axis=1)
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset = dataset.reset_index(drop=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False, dtype='float'), [0, 1, 2, 3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False, dtype='float'), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
quarter_index = 0
for i in range(X_train.shape[0]):
    if X_train[i, quarter_index] == 'Quarter1':
        X_train[i, quarter_index] = 1
    elif X_train[i, quarter_index] == 'Quarter2':
        X_train[i, quarter_index] = 2
    elif X_train[i, quarter_index] == 'Quarter3':
        X_train[i, quarter_index] = 3
    elif X_train[i, quarter_index] == 'Quarter4':
        X_train[i, quarter_index] = 4

for i in range(X_test.shape[0]):
    if X_test[i, quarter_index] == 'Quarter1':
        X_test[i, quarter_index] = 1
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```
